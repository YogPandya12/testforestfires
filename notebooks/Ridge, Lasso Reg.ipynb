```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
## Data cleaning
df[df.isnull().any(axis=1)]
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing null values
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
```

```python
## change the required columns as integer data type
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
### Changing the other columns to float datatpes
```

```python
obj=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in obj:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```

# Exploratory Data Analysis

```python
## drop day, month, year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
## Encoding of categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
## Categories in classses
df_copy['Classes'].value_counts()
```

```python
## Plot density for all features
plt.style.use('seaborn')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## ploting pie chart
classlabels=['fire','not fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
## Correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```
